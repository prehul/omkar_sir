```python
text=" ".join(cat.split()[1] for cat in df.category)
from wordcloud import WordCloud
word_cloud = WordCloud(height=2000, width=2000, background_color='black')
word_cloud = word_cloud.generate(text)
plt.imshow(word_cloud)
plt.title("Most common words in positive customer comments")
plt.axis('off')
plt.show()
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(height=2000, width=2000, background_color='white')
wordcloud = wordcloud.generate(" ".join(cat.split()[1] for cat in df.category))
plt.imshow(wordcloud)
plt.title("Most common words in positive customer comments")
plt.axis('off')
plt.show()
```
