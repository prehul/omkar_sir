**Steps to capture a video:**
    

- Use cv2.VideoCapture() to get a video capture object for the camera.

- Video is a combination of frames

- When we capture live video, it is like a infinite frames till we stop the video

- Set up an infinite while loop and use the read() method to read the frames using the above created object.

- Use cv2.imshow() method to show the frames in the video.

- Breaks the loop when the user clicks a specific key.

```python
#import cv2
# cv2.imread : Image read
# cv2.imshow : Image show
# cv2.imwrite: Image write
# cv2.VideoCapture(0)  # it will open the camera laptop. capture our videos
```

```python
# import the opencv library 
import cv2 
  
# define a video capture object 
vid = cv2.VideoCapture(0) 
  
while True:    
    # Capture the video frame 
    # by frame 
    ret, frame = vid.read() 
    # ret= boolean value
    # ret=True, the frame is captured
    # ret=False , the frame is not capture
    print(ret)
    # Display the resulting frame 
    cv2.imshow('frame', frame)   
    # the 'q' button is set as the 
    # quitting button you may use any 
    # desired button of your choice
    # 1 means one milli second
    # 1 and 0xFF its try identify an 8 digit key
    # ord('q') : ASCII 8 digits
    # so it will check are you entered any 8 digit key or not
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
        
# After the loop release the cap object 
vid.release() 
# Destroy all the windows 
cv2.destroyAllWindows() 
```

```python
# Importing all necessary libraries 
import cv2 
import os 

# Read the video from specified path 
cam = cv2.VideoCapture("los_angeles.mp4")                # Web cam 

try: 
    # creating a folder named data 
    if not os.path.exists('data'): 
        os.makedirs('data') 
# if not created then raise error 
except OSError: 
    print ('Error: Creating directory of data') 

# frame 
currentframe = 0

while True: 
    # reading from frame 
    ret,frame = cam.read() 
    if ret: 
        # if video is still left continue creating images 
        name = './data/frame' + str(currentframe) + '.jpg'  #  inside data folder
        print ('Creating...' + name) 
        # writing the extracted images 
        cv2.imwrite(name, frame) 
        # increasing counter so that it will 
        # show how many frames are created 
        currentframe += 1
    else: 
        break
        
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows() 
```

```python
import cv2 
  
cap = cv2.VideoCapture(0) 
  
while(True): 
      
    # Capture frames in the video 
    ret, frame = cap.read() 
    # describe the type of font to be used. 
    font = cv2.FONT_HERSHEY_SIMPLEX 
    # Use putText() method for 
    # inserting text on video    R=0,G=255,B=255
    cv2.putText(frame, 'Hello Omkar', (50, 50), font, 1,(0, 255, 255),2,cv2.LINE_4) 
  
    # Display the resulting frame 
    cv2.imshow('video', frame) 
  
    # creating 'q' as the quit  
    # button for the video 
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

# release the cap object 
cap.release() 
# close all windows 
cv2.destroyAllWindows() 
```

```python
import cv2
videopath = ("los_angeles.mp4")
cap = cv2.VideoCapture(videopath)
while True:
    ret, frame = cap.read()
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame, 'car video', (50, 50),
                font, 1, (0,255,255), 2, cv2.LINE_4)
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2 
  
def main(): 
    
    # reading the input 
    cap = cv2.VideoCapture(0) 
  
    output = cv2.VideoWriter( 
        "output.mp4", 
        cv2.VideoWriter_fourcc(*'MPEG'), 
        30, 
        (1080, 1920)) 
  
    while(True): 
        ret, frame = cap.read() 
        if(ret): 
              
            # adding rectangle on each frame 
            cv2.rectangle(frame, (100, 100), (500, 500), (0, 255, 0), 3) 
              
            # writing the new frame in output 
            output.write(frame) 
            cv2.imshow("output", frame) 
            if cv2.waitKey(1) & 0xFF == ord('s'): 
                break
  
    cv2.destroyAllWindows() 
    output.release() 
    cap.release() 
  
  
if __name__ == "__main__": 
    main()
```
