```python
#Routine to fix 
def fixColor(image):
    return(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
#Note the impact of the correct color scheme
plt.imshow(fixColor(image))
```

```python
#Translation matrix the first row moves pixels horizontally and the second row moves in vertically.
import numpy as np
M = np.float32([[1, 0, 250], [0, 1, 500]])
M

# 1 0  is horizontal axis
# 0 1  is vertical axis
```

```python
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape
[0]))
plt.imshow(shifted)
```

```python
#Changed the pixel directions to move image left and up
#[1, 0, -50] [row,column,number of pixels]
M = np.float32([[1, 0, -50],   # Horizontal move 
                [0, 1, -100]]) # Vertical move
shifted = cv2.warpAffine(image, 
                         M, 
                         (image.shape[1], 
                          image.shape[0]))
plt.imshow(shifted)
```

**Resizing**

```python
image.shape[0] # H
image.shape[1] # W
```

```python
r = 150.0 / image.shape[1]
r,image.shape[1]

# 150 is the width I have taken
# Now I want to know how much 150 is the percentage of original height
# 27%

# So I need to take 27% of Height also
# image.shape[0] * r
```

```python
dim = (150, int(image.shape[0] * r))
dim
```

```python
#Here we resize setting width to 150 and scale the height accordingly
# interploation is a method to downsampling the image
# resize the image === > shrinkage 
# INTER_AREA is the method
r = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * r))
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized)
```

```python
#Here we rezise setting height to 150 and scale the width accordingly
r = 150.0 / image.shape[0]
dim = ( int(image.shape[1] * r), 150)
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized)
```
