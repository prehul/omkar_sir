Hyperparameter tuning in logistic regression involves optimizing parameters that are not directly learned by the model during training. 

In logistic regression, the main hyperparameters to tune typically include:

**Regularization parameter (C):**

- Regularization helps prevent overfitting by penalizing large coefficients. 

- In scikit-learn's logistic regression implementation, this is controlled by the parameter C. 
    
- Lower values of C increase the regularization strength, while higher values decrease it.

**Penalty type:** 

- Logistic regression can use either L1 (Lasso) or L2 (Ridge) regularization penalties.
        
- L1 regularization tends to produce sparse solutions by shrinking some coefficients to zero,
        
- while L2 regularization tends to shrink all coefficients by a small amount. 
        
- This hyperparameter is controlled by the penalty parameter in scikit-learn.

**Maximum number of iterations:** 
        
- This is the maximum number of iterations taken for the solvers to converge. This is controlled by the max_iter parameter in scikit-learn. It's

important to set this parameter high enough to ensure convergence.




```python
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
## All together

#================================ Step-1: Get the base model===========================================
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()   # Base model
log_params=[keys for keys in log_reg.get_params()]
print("avialable parameters for model:",log_params)

print("==========================================================================================")

#================================ Step-2: Create a dictionary for params=================================
# Define hyperparameters and their values to be searched
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'max_iter': [100, 200, 300]
}

print("Processing parameters are:",param_grid)

print("===============================================================================================")

#============================== Step-3: Apply the Grid serach======================================
grid_search = GridSearchCV(log_reg,  # Base modl
                           param_grid, # params 
                           scoring='accuracy', # metric
                           cv=5,
                          verbose=True)

#============================= Step-4: fit the model with train data ===============================
grid_search.fit(X_train,y_train)

#============================= Step-5: Get the best parameters=====================================
best_params=grid_search.best_params_
best_score=grid_search.best_score_
print("best params:",best_params)
print("best train accuracy:",best_score)
```
